```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('foxnews-transcript.csv.gz')
df
```

```python
for i, r in df.iterrows():
    fn = r.html_file
    try:
        with open('html/' + fn) as f:
            soup = BeautifulSoup(f.read())
        texts = []
        elem = soup.find('p', {'class': 'speakable'})
        texts.extend(elem.get_text(strip=True, separator='\n').splitlines())
        for elem in elem.next_siblings:
            texts.extend(elem.get_text(strip=True, separator='\n').splitlines())
        with open('text/' + fn.replace('.html', '.txt'), 'w') as f:
            f.write('\n'.join(texts))
    except Exception as e:
        print(e, fn)
    if i % 100 == 0:
        print(i, fn)
    #break
```
