```python
import pandas as pd
import requests
impor time
```

```python
size = 30
start = 0
categories = []
while True:
    params = {'searchBy': 'categories',
              'values': 'fox-news/transcript',
              'size': size,
              'from': start
             }
    headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36"}
    r = requests.get('https://www.foxnews.com/api/article-search', params=params, headers=headers)
    print(start, r.status_code, r.url)
    j = r.json()
    print(len(j))
    if len(j) == 0:
        break
    categories.extend(j)
    start += size
    time.sleep(1)
```

```python
cat_df = pd.DataFrame(categories)
cat_df
```

```python
cat_df['cat_value'] = cat_df.category.apply(lambda c: 'fox-news' + c['url'].split('/category')[-1])
cat_df['cat_type'] = cat_df.category.apply(lambda c: c['name'])
cat_df
```

```python
cat_df[cat_df.cat_type != 'TRANSCRIPT']
```

```python
cat_df.cat_value.unique()
```

```python
data = []
for values in cat_df.cat_value.unique():
    print(values)
    if values.find('shows') == -1:
        continue
    start = 0
    size = 30
    while True:
        params = {'searchBy': 'tags',
                  'values': values,
                  'size': size,
                  'from': start,
                  'excludeBy': 'tags',
                  'excludeValues': ''
             }
        headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36"}
        r = requests.get('https://www.foxnews.com/api/article-search', params=params, headers=headers)
        print(start, r.status_code, r.url)
        j2 = r.json()
        print(len(j2))
        data.extend(j2)
        if len(j2) == 0:
            break
        start += size
        time.sleep(1)
    #break
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_excel('foxnews-transcript-urls.xlsx', index=False)
```
