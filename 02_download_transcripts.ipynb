```python
import pandas as pd
import requests
import time
import os
```

```python
df = pd.read_csv('foxnews-transcript-urls.csv.gz')
df
```

```python
df.drop_duplicates(subset=['url'], inplace=True)
df
```

```python
len(df.url.unique())
```

```python
df['html_file'] = df.url.apply(lambda c: c.split('/')[-1] + '.html')
df
```

```python
df[df.duplicated(subset=['html_file'])]
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
for i, r in df.iterrows():
    fn = 'html/' + r['html_file']
    if os.path.exists(fn):
        continue
    url = r['url']
    headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36"}
    try:
        rq = requests.get('https://www.foxnews.com/' + url, headers=headers)
        if rq.status_code == 200:
            with open(fn, 'wb') as f:
                f.write(rq.content)
        else:
            print('ERROR:', rq.status_code)
    except Exception as e:
        print('ERROR:', e)
    time.sleep(0.5)
    if i % 100 == 0:
        print(i, url)
    #break
```

```python
df.to_csv('foxnews-transcript.csv.gz', compression='gzip', index=False)
```
